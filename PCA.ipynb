```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
df = pd.read_csv('wine (1).csv')
df
```

```python
df['Type'].value_counts()
```

```python
df2=df.iloc[:,1:]
df2
```

```python
df2.shape
```

```python
df2.info()
```

```python
#converting data into array
arr=df2.values
arr
```

```python
#normalizing the numerical data
dfn=scale(df2)
dfn
```

# PCA Implementation

```python
pca=PCA(n_components=13)

dfpca=pca.fit_transform(dfn)
dfpca 
```

```python
#PCA components matrix
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
#cumulative variance of PCA
varc=np.cumsum(np.round(var,4)*100)
varc
```

```python
#variance plot
plt.plot(varc)
```

```python
#crating df
pca_df=pd.concat([df['Type'],pd.DataFrame(dfpca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
pca_df
```

```python
sns.pairplot(pca_df)
```

# 1) Hierarachical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#Dendogram
d= sch.dendrogram(sch.linkage(dfn,'complete'))
```

```python
#creating clusters
c = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
c
```

```python
y = pd.DataFrame(c.fit_predict(dfn),columns=['cluster_id'])
y['cluster_id'].value_counts()
```

```python
#adding clusters to dataset
df3 = df.copy()
df3['clusters_id']=c.labels_
df3
```

# 2) K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
WCSS=[]
for i in range(1,6):
    kmeans=KMeans(n_clusters=i, random_state=2)
    kmeans.fit(dfn)
    WCSS.append(kmeans.inertia_)
```

```python
#ploting K values vs wcss(elbow graph)
plt.plot(range(1,6),WCSS);
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

# Taking clusters K = 3

```python
c3 = KMeans(3,random_state=30).fit(dfn) 
c3
```

```python
c3.labels_
```

```python
#assigning clusters to dataset
dff = df.copy()
dff['clusters3_id']= c3.labels_
dff
```

```python
dff['clusters3_id'].value_counts()
```

Hence, we can conclude that the number of clusters are finalized as 3
